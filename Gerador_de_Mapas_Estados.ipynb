```python
data = {
    'Estado':['Rio Grande do Sul', 'Goiás', 'Pará', 'Santa Catarina', 'São Paulo'],
    'Valores':[80, 60, 40, 20, 10]
}
```

```python
#@title Baixar Dados
from geobr import read_state
estados = read_state()
```

```python
#@title Mapa
import pandas as pd
import plotly.express as px
from statistics import mean

df = pd.DataFrame(data)
df['Estado'] = df['Estado'].apply(lambda x: x.title())

estados = estados[estados['name_state'].isin(df['Estado'])]
estados = estados.rename(columns={'name_state': 'Estado'})
estados = pd.merge(estados, df, how='left', on = 'Estado')
estados.index = list(estados['Estado'])

fig = px.choropleth_mapbox(estados,
    geojson=estados.geometry,
    locations=estados.index,
    color="Valores",
    opacity=0.5,
    center={"lat": (((mean(list(estados.geometry.bounds.maxy))-mean(list(estados.geometry.bounds.miny)))/2)+mean(list(estados.geometry.bounds.miny)))
    , "lon": (((mean(list(estados.geometry.bounds.maxx))-mean(list(estados.geometry.bounds.minx)))/2)+mean(list(estados.geometry.bounds.minx)))},
    labels={'index':'Município'},
    mapbox_style="open-street-map",
    zoom=3,
)
fig.update_layout(margin=dict(l=1, r=1, t=1, b=1))
```
